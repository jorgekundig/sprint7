```text
obtener_datos():
    1. Pide el nombre del usuario mediante un input.
    2. Pide el teléfono del usuario mediente un input.
    3. Pide el nivel de experiencia del usuario mediante un input y lo convierte a valor tipo float.
    4. Pide el nivel de catalán del usuario mediante un input y lo convierte a valor tipo int.
    5. Pide el nivel de Excel del usuario mediante un input y lo convierte a valor tipo int.
    6. Pide el nivel de estudios del usuario mediante un input y lo convierte a valor tipo int.
    7. Guarda los valores anotados en un diccionario
    8. Devuelve el diccionario

````python
def obtener_datos():
    nombre = input("Escribe tu nombre completo: ")
    telefono = input("Introduce un teléfono de contacto: ")
    experiencia = float(input("¿Cuántos años tienes en la empresa? "))
    catalan = int(input("De acuerdo a la siguiente escala, ¿cuál es tu nivel de catalán?\n"
    "0 = sin conocimiento\n"
    "1 = hasta nivel B1\n"
    "2 = nivel B2\n"
    "3 = C1 o superior\n"
    "Anota tu nivel en número del 0 al 3: "))
    excel = int(input("De acuerdo a la siguiente escala, ¿cuál es tu dominio de Excel?\n"
    "0 = nivel básico\n"
    "1 = uso de funciones\n"
    "2 = tablas dinámicas\n"
    "3 = Visual Basic y Macros\n"
    "Anota tu nivel en número del 0 al 3: "))
    titulo = int(input("De acuerdo a la siguiente escala, ¿cuál es tu máximo grado de estudios?\n"
    "0 = ESO\n"
    "1 = FP\n"
    "2 = Universidad\n"
    "3 = Master o superior\n"
    "Anota tu nivel en número del 0 al 3 : "))
    candidato = {"Nombre":nombre, "Tel. de contacto":telefono, "Años Experiencia":experiencia, "Nivel Catalán":catalan, "Nivel Excel":excel, "Título Académico":titulo}
    return candidato
````

```text
validar_datos():
    1. Toma como parámetros los niveles de catalán, Excel, grado de estudios y años de experiencia.
        - Si los niveles de catalán, Excel y grado de estudios son números fuera de la escala 0-3, imprime un mensaje de error y devuelve un False
        - Si los años de experiencia son menores a 0 (número negativo), imprime un mensaje de error y devuelve un False
    2. Si los valores no cumplen estas condiciones (es decir, son valores correctos), devuelve un True

````python
def validar_datos(cat, exc, tit, exp):
    for nivel in (cat, exc, tit):
        if nivel < 0 or nivel > 3:
            print("Debe ser un número entero del 0 al 3")
            return False
    if exp < 0:
        print("Los años de experiencia no pueden ser negativos")
        return False
    return True
````

```text
calcular_puntuacion():
    1. Toma como parámetros los niveles de catalán, Excel, grado de estudios y años de experiencia.
    2. Los suma, multiplicando el valor de años de experiencia por 1.5
    3. Devuelve la puntuación total

````python
def calcular_puntuacion(cat, exc, tit, exp):
    punt_total = cat + exc + tit + (exp * 1.5)
    return punt_total
````

```text
clasificar_candidato():
    1. Toma como parámetro la puntuación total
        - Si es menor a 8, devuelve "No prioritario"
        - Si es mayor o igual a 8 y menor que 12, devuelve "Candidato/a adecuado/a"
        - Si es mayor a 12, devuelve "Prioridad Alta"

````python
def clasificar_candidato(punt_total):
    if punt_total < 8:
        return "No prioritario"
    elif punt_total <= 12:
        return "Candidato/a adecuado/a"
    else:
        return "Prioridad Alta"
````

```text
mostrar_individual():
    1. Toma como parámetro el diccionario del candidato
    2. Imprime su nombre y su puntuación al completar el registro

````python
def mostrar_individual(candidato):
    print(f"Hola, {candidato['Nombre']}, tu puntuación es de: {candidato['Puntuación Total']}")
````

```text
mostrar_informe():
    1. Toma como parámetro la lista de diccionarios de candidatos.
    2. Asigna valores numéricos a las categorías para poder ordenarlas según su prioridad, a modo diccionario clave:valor.
    3. Crea una lista que consiste en la lista de candidatos ordenada de acuerdo a la prioridad y a la puntuación total.
    4. Imprime el nombre, teléfono, puntuación total y categoría de cada candidato, de forma ordenada.

````python
def mostrar_informe(candidatos):
    prioridad = {"Prioridad Alta": 1, "Candidato/a adecuado/a": 2, "No prioritario": 3}
    candidatos_sort = sorted(
        candidatos, key=lambda x: (prioridad[x['Categoría']], -x['Puntuación Total'])
    )
    for candidato in candidatos_sort:
        print(f"{candidato['Nombre']}, {candidato['Tel. de contacto']}, {candidato['Puntuación Total']}, {candidato['Categoría']}")
````

```text
main_seleccion():
    1. Crea una lista vacía de candidatos donde se guardará el diccionario de cada candidato.
    2. Con un bucle while, limitamos a que se realicen 10 registros válidos. Al llegar al décimo, el bucle se detiene.
    3. La función obtener_datos() guarda en candidato.
    4. Mediante un condicional, si validar_datos() devuelve un True, se sigue ejecutando el resto de la función.
    5. La función calcular_puntuacion() almacena en punt_total.
    6. El valor de punt_total se añade al diccionario con la clave 'Puntuación Total'.
    7. La función clasificar_candidato almacena su valor en el diccionario con la clave 'Categoría'.
    8. Cada diccionario se añade a la lista candidatos.
    9. Se imprime la puntuación total de cada candidato al completar su registro.
    10. En caso de haber errores en el registro, se imprime un mensaje de error.
    11. Se imprime el número de candidatos registrados para llevar un control de que sean 10 registros válidos.
    12. Al completarse el bucle, mostrar_informe() muestra la lista ordenada de candidatos según su Categoría y Puntuación Total

````python
def main_seleccion():
    candidatos = []
    while len(candidatos) < 10:
        candidato = obtener_datos()
        if validar_datos(candidato['Nivel Catalán'], candidato['Nivel Excel'], candidato['Título Académico'], candidato['Años Experiencia']):
            punt_total = calcular_puntuacion(candidato['Nivel Catalán'], candidato['Nivel Excel'], candidato['Título Académico'], candidato['Años Experiencia'])
            candidato['Puntuación Total'] = punt_total
            categoria = clasificar_candidato(candidato['Puntuación Total'])
            candidato['Categoría'] = categoria
            candidatos.append(candidato)
            mostrar_individual(candidato)
        else:
            print("Registro inválido, intenta de nuevo")
        print(f"No. de candidatos registrados: {len(candidatos)}")
    print("")
    mostrar_informe(candidatos)
````

```text
obtener_segmento_y_factor():
    1. Incluye la escala de código de comportamiento a modo diccionario.
    2. La variable codigo almacena el valor correspondiente a la clave 'codigo_comportamiento'.
    3. La variable factor_codigo guarda el factor de código de comportamiento con un mapeo de diccionario con la escala que hemos creado.
    4. Se devuelve el factor_codigo.

````python
def obtener_segmento_y_factor(cliente):
    escala_comp = {0:0.1, 1:0.25, 2:0.50, 3:0.75, 4:0.9}
    codigo = cliente['codigo_comportamiento']
    factor_codigo = escala_comp[codigo]
    return factor_codigo
````

````python
def obtener_factor_edad(cliente):
    escala_edad = {'18-25':1.2, '26-40':1.0, '41-60':0.85, '60+':0.7}
    edad = cliente['grupo_edad']
    factor_edad = escala_edad[edad]
    return factor_edad
````

```text
asignar_mensaje():
    1. Incluye una tabla de correspondencia de mensajes, donde cada código de comportamiento recibe un mensaje de marketing.
    2. La variable codigo almacena el valor correspondiente a la clave 'codigo_comportamiento'
    3. La variable mensaje_mkt guarda el mensaje de marketing con un mapeo de diccionario con la escala que hemos creado.
    4. Se devuelve el mensaje_mkt.

````python
def asignar_mensaje(cliente):
    escala_mensajes = {
        0:'No te pierdas nuestro apartado de promociones y novedades para ver nuestros nuevos productos.',
        1:'Aún hay productos en tu canasta de compra, ¿quieres hacer el checkout?',
        2:'Estás a un paso de finalizar tu compra, ¿necesitas ayuda para hacer el pago?',
        3:'Tu canasta de compra te está esperando, ¡no pierdas tus productos!',
        4:'Tu canasta de compra te echa de menos, ¡finaliza tu compra!'
        }
    codigo = cliente['codigo_comportamiento']
    mensaje_mkt = escala_mensajes[codigo]
    return mensaje_mkt
````

```text
calcular_riesgo():
    1. La variable riesgo guarda el resultado de la multiplicación (redondeada) entre el factor_codigo y el factor_edad.
    2. Devuelve el riesgo

````python
def calcular_riesgo(factor_codigo, factor_edad):
    riesgo = round(factor_codigo * factor_edad, 2)
    return riesgo
````

```text
procesar_clientes():
    1. Orquesta a las funciones anteriores: obtener_segmento_y_factor(), obtener_factor_edad() y calcular_riesgo().
    2. Inicia un bucle para iterar los valores de cada cliente del diccionario de clientes.
    3. La variable factor_codigo almacena el llamado de la función obtener_y_factor.
    4. La variable factor_edad almacena el llamado de la función obtener_factor_edad.
    5. La variable riesgo almacena el llamado de la función calcular_riesgo
    6. Se incorpora al diccionario de cada cliente el valor de factor_codigo con la clave 'factor_codigo'.
    7. Se incorpora al diccionario de cada cliente el valor de factor_edad con la clave 'factor_edad'.
    8. Se incorpora al diccionario de cada cliente el valor de riesgo con la clave 'riesgo_final'.

````python
def procesar_clientes(clientes):
    for cliente in clientes.values():
        factor_codigo = obtener_segmento_y_factor(cliente)
        factor_edad = obtener_factor_edad(cliente)
        riesgo = calcular_riesgo(factor_codigo, factor_edad)
        mensaje_mkt = asignar_mensaje(cliente)
        cliente['factor_codigo'] = factor_codigo
        cliente['factor_edad'] = factor_edad
        cliente['riesgo_final'] = riesgo
        cliente['mensaje'] = mensaje_mkt
    return clientes
````

```text
separar_campos():
    1. Crea una lista vacía pacientes_split donde se guardarán los strings.
    2. Inicia el bucle para iterar por cada paciente.
    3. Con split("|"), se elimina el separador | y cada string queda separado.
    4. Se añade cada string a la lista vacía.
    5. Se devuelve la lista de pacientes, ya con los strings separados.

````python
def separar_campos(pacientes):
    pacientes_split = []
    for paciente in pacientes:
        campos = paciente.split("|")
        pacientes_split.append(campos)
    return pacientes_split
````

```text
convertir_edad():
    1. Crea una lista vacía pacientes_edad donde se guardará la lista de cada paciente.
    2. Inicia el bucle para iterar por cada paciente.
    3. Se asigna el índice 1 (segunda posición) al elemento en la segunda posición de la lista de cada paciente, convertido a número entero.
    4. Se añade cada lista a la lista vacía.
    5. Se devuelve la lista de pacientes, con la edad transformada en int.


````python
def convertir_edad(pacientes):
    pacientes_edad = []
    for paciente in pacientes:
        paciente[1] = int(paciente[1])
        pacientes_edad.append(paciente)
    return pacientes_edad
````

```text
separar_sintomas():
    1. Crea una lista vacía pacientes_sintomas donde se guardará la lista de cada paciente.
    2. Inicia el bucle para iterar por cada paciente.
    3. La variable nuevo_paciente guarda una lista con los sintomas de cada paciente.
    4. Se añade cada lista a la lista vacía.
    5. Se devuelve la lista de pacientes, con la edad transformada en int.

````python
def separar_sintomas(pacientes):
    pacientes_sintomas = []
    for paciente in pacientes:
        nuevo_paciente = [paciente[0], paciente[1], paciente[2]]
        pacientes_sintomas.append(nuevo_paciente)
    return pacientes_sintomas
````

```text
transformar_strings():
    1. Orquesta las funciones anteriores, tomando como parámetro la lista original de pacientes.
    2. La variable campos_separados guarda la lista después de llamar a separar_campos.
    3. La variable edad_entero guarda la lista después de llamar a convertir_edad.
    4. La variable sintomas_aislados guarda la lista después de llamar a separar_sintomas.
    5. Creamos una lista vacía, pacientes_dic.
    6. Inicia el bucle para iterar por cada paciente después de haber separado los síntomas.
    7. La variable nuevo_paciente transforma la lista de cada paciente en un diccionario, asignando las claves 'nombre', 'edad' y 'sintomas' a los valores correspondientes a los índices 0, 1 y 2 respectivamente.
    8. Se añade cada nuevo paciente a la lista vacía.
    9. Se devuelve la lista con el diccionario de cada paciente con los strings separados, dejando los sintomas dentro de una lista que funciona como valor a la clave 'sintomas'.

````python
def transformar_strings(pacientes):
    campos_separados = separar_campos(pacientes)
    edad_entero = convertir_edad(campos_separados)
    sintomas_aislados = separar_sintomas(edad_entero)
    pacientes_dic = []
    for paciente in sintomas_aislados:
        nuevo_paciente = {
            'nombre': paciente[0],
            'edad': paciente[1],
            'sintomas': paciente[2]
        }
        pacientes_dic.append(nuevo_paciente)
    return pacientes_dic
````

```text
dividir_comas():
    1. Inicia el bucle para iterar por cada paciente.
    2. La variable intermedia cadena_sintomas corresponde a los valores de la clave 'sintomas'.
    3. La variable sintomas_separados guarda los sintomas ya separados por coma, y sin espacios previos o posteriores, con los métodos split() y strip() respectivamente.
    4. Reasignamos esta variable sintomas_separados para que corresponda con la clave 'sintomas'.
    5. Devuelve la lista de diccionarios, con los sintomas separados en strings individuales.

````python
def dividir_comas(pacientes):
    for paciente in pacientes:
        cadena_sintomas = paciente['sintomas']
        sintomas_separados = cadena_sintomas.strip().split(",")
        paciente['sintomas'] = sintomas_separados
    return pacientes
````

```text
eliminar_espacios():
    1. Inicia el bucle para iterar por cada paciente.
    2. La variable nuevos_sintomas guarda una lista vacía.
    3. Inicia el bucle para iterar por cada síntoma.
    4. La variable intermedia sintoma_sin_espacios guarda cada sintoma. Mediante strip() y replace(), eliminamos espacios exteriores restantes y substituimos los espacios interiores (entre palabras del mismo síntoma) por una barra baja.
    5. Se guardan los síntomas ya transformados en la lista nuevos_sintomas.
    6. Para cada paciente, se corresponde la clave 'sintomas' con esta lista nueva.
    7. Se devuelve la lista de diccionarios, con los síntomas sin espacios y con barra baja para unificar formato original.

````python
def eliminar_espacios(pacientes):
    for paciente in pacientes:
        nuevos_sintomas = []
        for sintoma in paciente['sintomas']:
            sintoma_sin_espacios = sintoma.strip().replace(" ", "_")
            nuevos_sintomas.append(sintoma_sin_espacios)
        paciente['sintomas'] = nuevos_sintomas
    return pacientes
````

```text
minusculas():
    1. Inicia el bucle para iterar por cada paciente.
    2. La variable sintomas_minus guarda una lista vacía.
    3. Inicia el bucle para iterar por cada sintoma.
    4. La variable sintoma_min almacena cada sintoma después de pasarlo a minúsculas con lower().
    5. Se guardan los síntomas ya en minúsculas en la lista sintomas_minus.
    6. Para cada paciente, se corresponde la clave 'sintomas' con esta lista nueva.
    7. Se devuelve la lista de diccionarios, con los sintomas en minúsculas.

````python
def minusculas(pacientes):
    for paciente in pacientes:
        sintomas_minus = []
        for sintoma in paciente['sintomas']:
            sintoma_min = sintoma.lower()
            sintomas_minus.append(sintoma_min)
        paciente['sintomas'] = sintomas_minus
    return pacientes
````

```text
normalizar_sintomas():
    1. Orquesta las funciones anteriores, tomando como parámetro la lista de pacientes después de pasar por transformar_strings()
    2. La variable sintomas_comas guarda la lista de pacientes después de llamar a dividir_comas().
    3. La variable sintomas_sin_espacio guarda la lista de pacientes después de llamar a eliminar_espacios().
    4. La variable sintomas_minusculas guarda la lista de pacientes después de llamar a minusculas().
    5. Se devuelve la lista de diccionarios de pacientes ya transformada y procesada, con sintomas normalizados.

````python
def normalizar_sintomas(pacientes):
    sintomas_comas = dividir_comas(pacientes)
    sintomas_sin_espacio = eliminar_espacios(sintomas_comas)
    sintomas_minusculas = minusculas(sintomas_sin_espacio)
    return sintomas_minusculas
````

```text
asignar_riesgo():
    1. La variable niveles_riesgo guarda el diccionario donde se indica la correspondencia entre sintomas y nivel de riesgo (escala de riesgo).
    2. Inicia el bucle para iterar por cada paciente.
    3. La variable riesgos guarda una lista vacía.
    4. Inicia el bucle para iterar por cada síntoma.
    5. La variable valor_riesgo guarda la correspondencia entre el diccionario de riesgos y cada sintoma
    6. Se añade cada valor de riesgo a la lista vacía.
    7. Para cada paciente, se crea la clave 'valores_riesgo', correspondiente a la lista riesgos que hemos creado.
    8. Se devuelve la lista de pacientes, añadiendo una nueva clave 'valores_riesgo' con su respectiva lista de valores.

````python
def asignar_riesgo(pacientes):
    niveles_riesgo = {
        "fiebre_alta" : 3,
        "dificultad_respirar" : 5,
        "dolor_pecho" : 4,
        "tos" : 1,
        "fatiga" : 1,
        "dolor_cabeza" : 1
    }
    for paciente in pacientes:
        riesgos = []
        for sintoma in paciente['sintomas']:
            valor_riesgo = niveles_riesgo[sintoma]
            riesgos.append(valor_riesgo)
        paciente['valores_riesgo'] = riesgos
    return pacientes
````

```text
calcular_riesgo_total():
    1. Inicia el bucle para iterar por cada paciente.
    2. La variable riesgo_total guarda la suma de todos los valores correspondientes a la clave 'valores_riesgo' de cada paciente.
    3. Se asigna una nueva clave 'riesgo_total', que corresponde al valor de riesgo_total calculado en el paso previo.
    4. Se devuelve la lista de pacientes, con el riesgo_total añadido.

````python
def calcular_riesgo_total(pacientes):
    for paciente in pacientes:
        riesgo_total = sum(paciente['valores_riesgo'])
        paciente['riesgo_total'] = riesgo_total
    return pacientes
````

```text

definir_longitud():

```text
extraer_obligatorios():
    1. Toma como parámetros el diccionario de colección de caracteres y los parámetros opcionales (obligatorios para el ejercicio, que indican el tipo de caracteres que incluirá la contraseña).
    2. La variable parametros guarda un diccionario de parametros, que relaciona cada clave 'tipo' (tipo de caracter) con el valor booleano correspondiente True o False.
    3. La variable carac_obligatorios inicia una lista vacía, que irá guardando los caracteres.
    4. Inicia el bucle for para iterar por clave, valor el diccionario de caracteres, donde tipo representa el nombre de la categoría y caracteres su colección de caracteres asociada.
    5. Si el parámetro corresponde al tipo (es decir, la categoría está activada), la función choice del generador default_rng() de numpy extrae un valor aleatorio de ese tipo y se guarda en la variable caracter.
    7. Si la condición previa se cumplió, se añade el caracter extraido a la lista carac_obligatorios con el método append(), para garantizar tener al menos 1 caracter de cada tipo activado como parámetro.
    8. Se devuelve la lista de caracteres obligatorios.


Esta función utiliza dos diccionarios relacionados por las mismas claves (mayusculas, minusculas, numeros y signos). Gracias a ello, el programa puede comprobar qué categorías están activadas y seleccionar automáticamente caracteres de las colecciones correspondientes sin necesidad de escribir un bloque if independiente para cada tipo de carácter.

````python
def extraer_obligatorios(col_caracteres, mayusculas=True, minusculas=True, numeros=True, signos=False):
    parametros = {
        'mayusculas':mayusculas,
        'minusculas':minusculas,
        'numeros':numeros,
        'signos':signos
        }
    carac_obligatorios = []
    for tipo, caracteres in col_caracteres.items():
        if parametros[tipo]:
            caracter = rng.choice(list(caracteres))
            carac_obligatorios.append(caracter)
    return carac_obligatorios
````

```text
completar_pswd():
    1. Toma como parámetros el diccionario de colección de caracteres, la longitud de contraseña validada y los parámetros opcionales.
    2. La variable parametros guarda un diccionario de parametros, que relaciona cada clave 'tipo' (tipo de caracter) con el valor booleano correspondiente True o False.
    3. La variable col_aprobada inicia una lista vacía, donde se irán guardando los caracteres adicionales de las categorías activadas.
    4. Inicia el bucle for para iterar por clave, valor el diccionario de caracteres, donde tipo representa el nombre de la categoría y caracteres su colección de caracteres asociada.
    5. Si el parámetro corresponde al tipo (es decir, la categoría está activada), la lista suma los caracteres de la categoría activada (concatena su cadena de caracteres).
    6. La variable lista_col_aprobada convierte esa cadena en una lista de caracteres individuales para poder ser utilizada por la función choice() de numpy.
    7. La variable num_resto_caracteres calcula cuántos caracteres faltan para alcanzar la longitud solicitada por el usuario, restando la cantidad de caracteres obligatorios ya generados.
    8. La variable relleno almacena los caracteres aleatorios seleccionados con la función choice del generador default_rng() de numpy, la cantidad ide ellos calculada en el paso anterior.
    8. La variable pswd_completo guarda la concatenación de los caracteres obligatorios + los caracteres "de relleno".
    8. Se devuelve la contraseña completa en una lista pswd_completo, con la cantidad de caracteres solicitada y de las categorías activadas.

````python
def completar_pswd(col_caracteres, long_pswd_ok, carac_obligatorios, mayusculas=True, minusculas=True, numeros=True, signos=False):
    parametros = {
        'mayusculas':mayusculas,
        'minusculas':minusculas,
        'numeros':numeros,
        'signos':signos
        }
    col_aprobada = []
    for tipo, caracteres in col_caracteres.items():
        if parametros[tipo]:
            col_aprobada += caracteres
    lista_col_aprobada = list(col_aprobada)
    num_resto_caracteres = long_pswd_ok - len(carac_obligatorios)
    relleno = rng.choice(lista_col_aprobada, size = num_resto_caracteres)
    pswd_completo = carac_obligatorios + list(relleno)
    return pswd_completo
````

````python
def mezclar_pswd(pswd_completo):
    rng.shuffle(pswd_completo)
    return pswd_completo
````

generar_pswd():

    3. La variable carac_obligatorios guarda la lista de caracteres mínimos obligatorios (uno por categoría) y toma como parámetros la colección de caracteres y los parámetros opcionales.
    4. La variable pswd_completo guarda la lista de caracteres completa, resultado de pasar como parámetros la colección de caracteres, la longitud validada y los parámetros opcionales a la función completar_pswd().
    5. La función pswd_mezclado guarda la lista de caracteres completa con el orden alterado, después de pasar por la función mezclar_pswd().
    5. Se devuelve la lista de caracteres, con el orden alterado aleatoriamente.

````python
def generar_pswd(long_pswd_ok, mayusculas=True, minusculas=True, numeros=True, signos=False):
    col_caracteres = coleccion_caracteres()
    carac_obligatorios = extraer_obligatorios(col_caracteres, mayusculas, minusculas, numeros, signos)
    pswd_completo = completar_pswd(col_caracteres, long_pswd_ok, carac_obligatorios, mayusculas, minusculas, numeros, signos)
    pswd_mezclado = mezclar_pswd(pswd_completo)
    return pswd_mezclado
````

```text
devolver_pswd():
1. Toma como parámetro la lista de caracteres con el orden alterado.
2. Con el método join(), se concatenan los caracteres en una cadena de texto y se almacenan en la variable pswd_final.
3. Se devuelve la contraseña final, con la longitud solicitada y los tipos de caracteres activados.

````python
def devolver_pswd(pswd_mezclado):
    pswd_final = "".join(pswd_mezclado)
    return pswd_final
````

```text
main_contrasenas_seguras():
1. Toma los parámetros opcionales para activar el tipo de caracteres que el programa tomará para crear la contraseña.
2. Inicia un condicioal if not any, para forzar al usuario a activar al menos un tipo de caracter, ya que de lo contrario se devolvería una lista vacía que no podría ser transformada en una cadena de texto.
3. La variable long_pswd_ok guarda la longitud, después de haber sido solicitada y validada por definir_longitud().
4. Se imprime un mensaje para confirmar la longitud de la contraseña a crear.

6. La variable pswd_final guarda la contraseña después de haber sido convertida a cadena de texto por la función devolver_pswd().
7. Se imprime un mensaje mostrando la contraseña creada y se devuelve.

````python
def main_contrasenas_seguras(mayusculas=True, minusculas=True, numeros=True, signos=False):
    if not any([mayusculas, minusculas, numeros, signos]):
        print("Debes activar al menos un tipo de carácter.")
        return
    long_pswd_ok = definir_longitud()
    print(f"Longitud de contraseña elegida: {long_pswd_ok} caracteres. ")
    pswd_mezclado = generar_pswd(long_pswd_ok, mayusculas, minusculas, numeros, signos)
    pswd_final = devolver_pswd(pswd_mezclado)
    print(f"Tu nueva contraseña es: {pswd_final}")
    return pswd_final
````

````python
# Previamente, guardamos la ruta del fichero en una variable fichero_futbol e importampos pandas
import pandas as pd

from google.colab import files

print("Seleccione el archivo con los partidos: ")
files.upload()

ruta_fichero = "historic partits.txt"
````

```text
leer_fichero(ruta_fichero)
    1. Toma como parámetro la ruta del fichero txt.
    2. La función open abre el fichero y lo guarda en la variable archivo.
    3. La variable lineas guarda una lista vacía, que se irá llenando con cada línea del fichero.
    4. Inicia el bucle para iterar por cada línea del fichero (archivo).
    5. Se añade cada línea al fichero y se eliminan los espacios en blanco iniciales y finales con strip().
    6. Devuelve la lista de líneas, que son cadenas de texto.


````python
def leer_fichero(ruta_fichero):
    with open(ruta_fichero, 'r', encoding='utf-8') as archivo:
        lineas = []
        for linea in archivo:
            lineas.append(linea.strip())
    return lineas
````

```text
normalizar_lineas(lineas)
    1. Toma como parámetro la lista de cadenas de texto.
    2. La variable lineas_limpias guarda una lista vacía, que se irá llenando con cada línea del fichero .
    3. Inicia el bucle para iterar por cada cadena de texto de la lista.
    4. Se añade cada línea al fichero y se eliminan los espacios.
    5. Devuelve la lista de líneas, que son cadenas de texto con los equipos y marcador separados por un espacio.

````python
def normalizar_lineas(lineas):
    lineas_limpias = []
    for linea in lineas:
        linea_limpia = " ".join(linea.split())
        lineas_limpias.append(linea_limpia)
    return lineas_limpias
````

```text
separar_partidos(lineas_limpias)
    1. Toma como parámetro la lista de cadenas de texto, después de ser procesada en la función previa.
    2. La variable partidos_divididos guarda una lista vacía, que se irá llenando con cada linea del fichero.
    3. Inicia el bucle para iterar por cada cadena de texto de la lista
    5. La variable palabras guarda una lista con las palabras de cada partido/línea ('Equipo', A', 'marcador', 'Equipo', B').
    6. Se añade cada lista de elementos (palabras) a la lista partidos_divididos con append().
    6. Devuelve la lista de listas, donde cada lista contiene cada palabra como elementos separados.

````python
def separar_partidos(lineas_limpias):
    partidos_divididos = []
    for linea in lineas_limpias:
        palabras = linea.split()
        partidos_divididos.append(palabras)
    return partidos_divididos
````

```text
extraer_partidos(partidos_divididos)
    1. Toma como parámetro la lista de listas de partidos.
    2. La variable lista_partidos guarda una lista vacía, que se irá llenando con los partidos y sus respectivos datos.
    3. Inicia el bucle para iterar por cada "partido" (lista) de partidos_divididos.
    4. La variable dic_partido guarda un diccionario vacío, que se irá llenando con los datos un partido.
    5. Para encontrar el índice del marcador, que varía de acuerdo a la cantidad de palabras de la lista, usamos enumerate, con el condicional de que la palabra (en este caso el marcador) incluya un guión.
    6. Guardamos el índice de este marcador en la variable indice_marcador, que servirá como base para recorrer la lista de palabras con slicers.
    7. La variable marcador guarda el valor correspondiente al índice ('2-1', por ejemplo).
    8. La variable marcador_separado guarda una pequeña lista que incluye los goles del local y del visitante (índices 0 y 1, pues se elimina el guión) como elementos separados.
    9. La variable goles_local guarda el valor del índice 0 del marcador_separado y lo convierte a número entero.
    10. La variable goles_visitante guarda el valor del índice 1 del marcador_separado y lo convierte a número entero.
    11. La variable local guarda el nombre del equipo local, que serían todos los elementos que están antes del indice_marcador, unidos con " ".join para poder mantener espacios en los nombres de equipo de más de una palabra (ie. pasa de 'FC', 'Barcelona' a 'FC Barcelona').
    12. La variable visitante hace lo mismo, para el equipo visitante (los elementos después del indice_marcador).
    13. Con update, el diccionario de partido (dic_partido) va añadiendo los nombres de los equipos y los goles de local y de visitante.
    14. Se añade cada diccionario de partido a la lista de partidos (lista_partidos) creada al inicio de la función.
    15. Devuelve la lista de partidos.

````python
def extraer_partidos(partidos_divididos):
    lista_partidos = []
    for partido in partidos_divididos:
        dic_partido = {}
        indice_marcador = [i for i, palabra in enumerate(partido) if '-' in palabra]
        indice_marcador = indice_marcador[0]
        marcador = partido[indice_marcador]
        marcador_separado = marcador.split("-")
        goles_local = int(marcador_separado[0])
        goles_visitante = int(marcador_separado[1])
        local = " ".join(partido[:indice_marcador])
        visitante = " ".join(partido[indice_marcador + 1:])
        dic_partido.update({
            'local':local,
            'visitante':visitante,
            'goles_local':goles_local,
            'goles_visitante':goles_visitante})
        lista_partidos.append(dic_partido)
    return lista_partidos
````

```text
actualizar_clasificacion(lista_partidos)
    1. Toma como parámetro la lista de partidos.
    2. La variable tabla_general guarda un diccionario de equivalencias de resultados, en donde la victoria da 3 puntos, el empate 1 y la derrota 0.
    3. Inicia el bucle para recorrer cada partido de la lista.
    4. Para cada partido, las variables loc, vis, goles_loc y goles_vis corresponden al equipo local, el visitante y sus respectivos goles.
    5. Se inicializan los equipos: si el equipo local o el visitante (loc, vis) no están ya añadidos a la tabla_general, se añaden sus datos, con las estadísticas iniciando en 0.
    6. Para el equipo local de cada partido, sus goles a favor suman a la variable goles_loc, mientras que sus goles en contra suman a la variable goles_vis.
    7. Hacemos el mismo procedimiento para el equipo visitante.
    8. Si los goles de local son más que los de visitante, el equipo local suma un partido ganado y 3 puntos, según la tabla de equivalencias. El equipo visitante suma un partido perdido y los puntos correspondientes (0).
    9. Si son más los goles de visitante que de local, cambia la lógica: el equipo local suma un partido perdido y 0 puntos, mientras que el visitante obtiene un partido ganado y 3 puntos.
    10. Si hay igualdad de goles (else), ambos equipos suman un empate y 1 punto.
    11. Adicionalmente, para cada equipo en la tabla general, calculamos la diferencia de goles (goles a favor - goles en contra).
    12. Devuelve la tabla_general, un diccionario donde cada equipo funge como clave y su valor es un diccionario de resultados (puntos, ganados, empates, etc).

````python
def actualizar_clasificacion(lista_partidos):
    tabla_general = {}
    tabla_eq_puntos = {
    'victoria':3,
    'empate':1,
    'derrota':0
    }
    for partido in lista_partidos:
        loc = partido['local']
        vis = partido['visitante']
        goles_loc = partido['goles_local']
        goles_vis = partido['goles_visitante']
        if loc not in tabla_general:
            tabla_general[loc] = {'equipo':loc, 'puntos':0, 'ganados':0, 'empates':0, 'perdidos':0, 'goles_favor':0, 'goles_contra':0, 'dif_goles':0}
        if vis not in tabla_general:
            tabla_general[vis] = {'equipo':vis, 'puntos':0, 'ganados':0, 'empates':0, 'perdidos':0, 'goles_favor':0, 'goles_contra':0, 'dif_goles':0}
        tabla_general[loc]['goles_favor'] += goles_loc
        tabla_general[loc]['goles_contra'] += goles_vis
        tabla_general[vis]['goles_favor'] += goles_vis
        tabla_general[vis]['goles_contra'] += goles_loc
        if goles_loc > goles_vis:
            tabla_general[loc]['ganados'] += 1
            tabla_general[loc]['puntos'] += tabla_eq_puntos['victoria']
            tabla_general[vis]['perdidos'] += 1
            tabla_general[vis]['puntos'] += tabla_eq_puntos['derrota']
        elif goles_loc < goles_vis:
            tabla_general[loc]['perdidos'] += 1
            tabla_general[loc]['puntos'] += tabla_eq_puntos['derrota']
            tabla_general[vis]['ganados'] += 1
            tabla_general[vis]['puntos'] += tabla_eq_puntos['victoria']
        else:
            tabla_general[loc]['empates'] += 1
            tabla_general[loc]['puntos'] += tabla_eq_puntos['empate']
            tabla_general[vis]['empates'] += 1
            tabla_general[vis]['puntos'] += tabla_eq_puntos['empate']
    for equipo in tabla_general.values():
        equipo['dif_goles'] = equipo['goles_favor'] - equipo['goles_contra']
    return tabla_general
````

```text
mostrar_dataframe(tabla_general):
    1. Toma como parámetro el diccionario de diccionarios de equipos (tabla_general).
    2. La variable lista_tabla_general guarda el diccionario de diccionarios, transformado en una lista de diccionarios.
    3. La variable df_tabla_general guarda el diccionario transformado en dataframe, con el índice reseteado.
    4. Se devuelve el dataframe con los datos de cada equipo.

````python
def mostrar_dataframe(tabla_general):
    lista_tabla_general = list(tabla_general.values())
    df_tabla_general = pd.DataFrame(lista_tabla_general).reset_index(drop=True)
    return df_tabla_general
````

```text
ordenar_clasificacion(df_tabla_general):
    1. La variable equipo_goleador guarda el valor (nombre) correspondiente al equipo con el máximo valor de goles a favor.
    2. La variable goles_equipo_goleador guarda el valor (num. de goles) correspondiente al equipo con el máximo valor de goles a favor.
    3. Se imprimen estos datos.
    4. La variable equipo_goleado guarda el valor (nombre) correspondiente al equipo con el máximo valor de goles en contra.
    5. La variable goles_equipo_goleado guarda el valor (num. de goles) correspondiente al equipo con el máximo valor de goles en contra.
    6. Se imprimen estos datos.
    7. La variable df_tabla_ordenada guarda el dataframe después de ser reordenado, tomando como criterios los puntos, diferencia de goles y goles a favor (en esa jerarquía), todos de mayor a menor.
    8. Se imprime el dataframe.



````python
def ordenar_clasificacion(df_tabla_general):
    equipo_goleador = df_tabla_general.loc[df_tabla_general['goles_favor'].idxmax(), 'equipo']
    goles_equipo_goleador = df_tabla_general.loc[df_tabla_general['goles_favor'].idxmax(), 'goles_favor']
    print(f"Equipo más goleador: {equipo_goleador} - {goles_equipo_goleador} goles a favor")
    equipo_goleado = df_tabla_general.loc[df_tabla_general['goles_contra'].idxmax(), 'equipo']
    goles_equipo_goleado = df_tabla_general.loc[df_tabla_general['goles_contra'].idxmax(), 'goles_contra']
    print(f"Equipo más goleado: {equipo_goleado} - {goles_equipo_goleado} goles en contra")
    print("Clasificación final: ")
    print("")
    df_tabla_ordenada = df_tabla_general.sort_values(by=['puntos', 'dif_goles', 'goles_favor'], ascending=[False, False, False]).reset_index(drop=True)
    print(df_tabla_ordenada)
````

```text
main_datos_deportivos(ruta_fichero):
    1. Toma como parámetro la ruta del fichero txt.
    2. La variable lineas guarda el llamado de leer_fichero() -> una lista de cadenas de texto sin espacios exteriores.
    3. La variable lineas_limpias guarda el llamado de normalizar_lineas() -> una lista de cadenas de texto sin tabuladores, con espacio entre los elementos/palabras.
    4. La variable partidos_divididos guarda el llamado de separar_partidos() -> una lista de partidos, con cada elemento/palabra por separado.
    5. La variable lista_partidos guarda el llamado de extraer_partidos() -> una lista de partidos, donde cada partido es un diccionario que incluye equipo local, visitante, goles_local y goles_visitante.
    6. La variable tabla_general guarda el llamado de actualizar_clasificacion() -> un diccionario de equipos, donde cada equipo es un diccionario que incluye puntos, ganados, empates, derrotas, goles a favor, goles en contra y diferencia de goles.
    7. La variable df_tabla_general guarda el llamado de mostrar_dataframe() -> un dataframe con los datos de cada equipo.
    8. La variable df_tabla_ordenada guarda el llamado de ordenar_clasificacion() -> un dataframe ordenado por puntos - diferencia de goles - goles a favor, junto con el equipo más goleador y el más goleado.
    9. Se devuelve esta variable.

````python
def main_datos_deportivos(ruta_fichero):
    lineas = leer_fichero(ruta_fichero)
    lineas_limpias = normalizar_lineas(lineas)
    partidos_divididos = separar_partidos(lineas_limpias)
    lista_partidos = extraer_partidos(partidos_divididos)
    tabla_general = actualizar_clasificacion(lista_partidos)
    df_tabla_general = mostrar_dataframe(tabla_general)
    df_tabla_ordenada = ordenar_clasificacion(df_tabla_general)
    return df_tabla_ordenada
````
